## Advanced Lane Finding Project

The goals / steps of this project are the following:

* Compute the camera calibration matrix and distortion coefficients given a set of chessboard images.
* Apply a distortion correction to raw images.
* Use color transforms, gradients, etc., to create a thresholded binary image.
* Apply a perspective transform to rectify binary image ("birds-eye view").
* Detect lane pixels and fit to find the lane boundary.
* Determine the curvature of the lane and vehicle position with respect to center.
* Warp the detected lane boundaries back onto the original image.
* Output visual display of the lane boundaries and numerical estimation of lane curvature and vehicle position.


## Camera Calibration

*1. Camera matrix and distortion coefficients.*

The code for this step is contained in the below code cell. I start by preparing "object points", which will be the (x, y, z) coordinates of the chessboard corners in the world. Here I am assuming the chessboard is fixed on the (x, y) plane at z=0, such that the object points are the same for each calibration image.  Thus, `objp` is just a replicated array of coordinates, and `objpoints` will be appended with a copy of it every time I successfully detect all chessboard corners in a test image.  `imgpoints` will be appended with the (x, y) pixel position of each of the corners in the image plane with each successful chessboard detection.  

I then used the output `objpoints` and `imgpoints` to compute the camera calibration and distortion coefficients using the `cv2.calibrateCamera()` function.  I applied this distortion correction to the test image using the `cv2.undistort()` function and obtained the output shown below: 


```python
import cv2
import glob
import pickle
import random
import numpy as np
import matplotlib.image as mpimg
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
from scipy import interpolate
from scipy.signal import gaussian

%matplotlib inline

# Prepare object points, like (0,0,0), (1,0,0), (2,0,0) ....,(6,5,0)
objp = np.zeros((6*9,3),np.float32)
objp[:,:2] = np.mgrid[0:9,0:6].T.reshape(-1,2)

# Arrays to store object points and image points from all the images.
objpoints = []
imgpoints = []

# Make a list of calibration images
images = glob.glob('./camera_cal/calibration*.jpg')

# Sort list based on numerical index
images = sorted(images, key=lambda string: int(string[-6:-4]))
imgnum = [string[-6:-4] for string in images]

# Step through the list and search for chessboard corners
for idx, fname in zip(imgnum,images):
    img = cv2.imread(fname)
    gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

    # Find the chessboard corners
    ret, corners = cv2.findChessboardCorners(gray, (9,6), None)
    
    # If found, add object points, image points
    if ret == True:
        print('working on', fname)
        objpoints.append(objp)
        imgpoints.append(corners)
        
        # Draw and save images with detected corners
        cv2.drawChessboardCorners(img, (9,6), corners, ret)
        write_name = './camera_cal/'+'corners_found'+idx+'.jpg'
        cv2.imwrite(write_name, img)

img = cv2.imread('./camera_cal/calibration02.jpg')
img_size = (img.shape[1], img.shape[0])

global ret, mtx, dist

# Calculate camera matrix and distortion coefficients
ret, mtx, dist, rvecs, tvecs = cv2.calibrateCamera(objpoints, imgpoints, img_size, None, None)

img_matched = cv2.imread('./camera_cal/corners_found02.jpg')

img_distorted = cv2.imread('./camera_cal/calibration01.jpg')

# Apply distortion correction to test image
img_undistorted = cv2.undistort(img_distorted,mtx,dist,None,mtx)

gs = gridspec.GridSpec(2, 2)
gs.update(wspace=0.2, hspace=0.2) 
plt.figure(figsize=(16,9))

plt.subplot(2,2,1)
plt.imshow(img.squeeze())
plt.title('Calibration Image')

plt.subplot(2,2,2)
plt.imshow(img_matched.squeeze())
plt.title('Corners Found')

plt.subplot(2,2,3)
plt.imshow(img_distorted.squeeze())
plt.title('Distorted Image')

plt.subplot(2,2,4)
plt.imshow(img_undistorted.squeeze())
plt.title('Undistorted Image')

plt.show()
    
```


## Pipeline (Single Images)


The following two code cells contain the pipeline to process single images:


```python
class Line():
# Defines a class for each lane to detect lane centroids, filter lanes and store lane parameters in each frame        
    
    def __init__(self,nature):
    # initializes lane parameters/sets hyperparameters    
        self.type = nature
        self.kernel_width = 20
        self.num_windows = 9
        self.margin_pct = 0.05            
        self.ym_per_pix = 30/720
        self.xm_per_pix = 3.7/700
        self.reset = False
        
        self.detected = False  
        self.current_x = None  
        self.current_y = None
        self.false_detection_count = 0
        self.max_false = 5              # 25
        
        self.recent_fit = []      
        self.avg_fit = None
        self.current_fit = None 
        self.diff_fit = np.array([0,0,0],dtype='float')
         
        self.recent_xfitted = [] 
        self.avg_xfitted = None
        self.current_xfitted = None
        self.diff_xfitted = None
        self.yfitted = None
        
        self.curvature = None 
        self.current_slope = None
        
        self.prev_frames = 5    
        self.xfitted_error_pct = 0.10
        self.fit_error_pct = 0.25
        self.low_winthresh = 50
        self.high_winthresh = 100
        self.min_pixels = 10  
        self.warp_size = None 
        
    def fit_quadratic(self, x_pixels = None, y_pixels = None, y_fitted = None, poly_fit = None):
    # fits quadratic curve based on least squared error given y points / calculates x points given curve    
        if poly_fit is None:
            poly_fit = np.polyfit(y_pixels, x_pixels, 2)
        poly_fitx = poly_fit[0]*y_fitted*y_fitted + poly_fit[1]*y_fitted + poly_fit[2]
        poly_fitx = np.array(poly_fitx, np.int32)
        return poly_fit, poly_fitx
    
    def find_window_centroids(self, warped):
    # finds lane centroids in warped binary image    
        self.warp_size = warped.shape
                                         
        kernel_width = self.kernel_width
        window_height = int(warped.shape[0]/self.num_windows)      
        margin = self.margin_pct*self.warp_size[1]

        window_centroids = []
        kernel = gaussian(kernel_width, std=kernel_width/4, sym=True) # defines a gaussian window for convolution
        
        # bottom centroid is calculated for respective left/right lane
        if self.type == 'Left':
            lower_sum = np.sum(warped[int(warped.shape[0]-2*window_height):,:int(warped.shape[1]/2)], axis = 0)
            horizontal_offset = 0 
        elif self.type == 'Right':   
            lower_sum = np.sum(warped[int(warped.shape[0]-2*window_height):,int(warped.shape[1]/2):], axis = 0)
            horizontal_offset = int(warped.shape[1]/2)
           
       
        convolved = np.convolve(kernel,lower_sum,mode='same') 
        
        # constraints must be satisfied by centroid peak and window for robust detection
        if np.amax(convolved) >= self.high_winthresh\
        and len(np.where(convolved>self.low_winthresh)[0]) >=self.min_pixels\
        or self.avg_fit is None or self.reset is True:
             # convolution is done on the sum to find maximum(peaks)
            xcenter = np.argmax(convolved)+horizontal_offset
        else: # used in videos
            xcenter = self.avg_xfitted[-1]
                
        ycenter = warped.shape[0]                                                                                                                      
        window_centroids.append((xcenter,ycenter))   
         
        # centroids are found upwards progressively for vertical windows having specified width    
        for level in range(1,self.num_windows+1):
                
            ycenter = max(int(warped.shape[0] - level*window_height),0)                        
            image_layer = np.sum(warped[int(ycenter):int(window_centroids[-1][1]),:], axis = 0)
                                                       
            # finds centroids in region around previous high confidence detections if available
            if self.detected is True and self.reset is not True:                        
                _, xcenter_prev = self.fit_quadratic(y_fitted=ycenter, poly_fit=self.avg_fit)
                xmin_index = int(max(xcenter_prev-margin,0))
                xmax_index = int(min(xcenter_prev+margin,warped.shape[1]))
            else:
                xmin_index = int(max(xcenter-margin,0))
                xmax_index = int(min(xcenter+margin,warped.shape[1]))    
                
            convolved = np.convolve(kernel, image_layer[xmin_index:xmax_index+1], mode='same')
                
            if np.amax(convolved) >= self.high_winthresh\
            and len(np.where(convolved>self.low_winthresh)[0]) >=self.min_pixels:
                xcenter = np.argmax(convolved)+xmin_index 
            
            window_centroids.append((xcenter,ycenter))
                                       
        return window_centroids

    
    def tracking_operations(self, x_pixels, y_pixels, y_fitted):
    # performs tracking operations - filtering, averaging and reset for lanes      
        self.yfitted = y_fitted
        # if no lanes are detected in current frame previous frames are used if available
        if len(x_pixels)!=0 and len(y_pixels)!=0:
            # detects current lane candidate
            self.current_x = x_pixels
            self.current_y = y_pixels
            self.current_fit, self.current_xfitted = self.fit_quadratic(x_pixels=x_pixels,y_pixels=y_pixels,y_fitted=y_fitted)
            self.current_slope = 2*self.current_fit[0]*y_fitted + self.current_fit[1]
            # checks if initial frame or reset must be done
            if not(self.avg_fit is None) and self.false_detection_count < self.max_false:                
                self.reset = False
                self.diff_fit = np.absolute(self.current_fit - self.avg_fit)
                cond1 = self.diff_fit[-1] <= self.fit_error_pct*self.warp_size[1]
                self.diff_xfitted = np.absolute(self.current_xfitted - self.avg_xfitted) 
                cond2 = self.diff_xfitted <= self.xfitted_error_pct*self.warp_size[1]
                # applies filtering based on closeness to average values; false detection count is modified
                if np.all(cond1) and np.all(cond2):
                    self.detected = True
                    if self.false_detection_count > 0:
                        self.false_detection_count -= 1
                    self.recent_fit.append(self.current_fit)      
                    self.recent_xfitted.append(self.current_xfitted)
                else:
                    self.detected = False
                    self.false_detection_count += 1
                    print('filtered!')
            else:
                # tracking is reset in the case of numerous false detections
                if not(self.avg_fit is None):
                    #reset frame is validated against most recent detection 
                    most_recent_fit = self.recent_fit[-1]
                    most_recent_xfitted = self.recent_xfitted[-1]
                    self.detected = False
                    self.false_detection_count = 0
                    self.reset = True
                    print('reset!!')
                    self.recent_fit = []
                    self.recent_xfitted = []
                    self.diff_fit = np.absolute(self.current_fit - most_recent_fit)
                    cond1 = self.diff_fit[-1] <= self.fit_error_pct*self.warp_size[1]
                    self.diff_xfitted = np.absolute(self.current_xfitted - most_recent_xfitted) 
                    cond2 = self.diff_xfitted <= self.xfitted_error_pct*self.warp_size[1]
                    if np.all(cond1) and np.all(cond2):
                        self.recent_fit.append(self.current_fit)
                        self.recent_xfitted.append(self.current_xfitted)
                    else:
                        self.recent_fit.append(most_recent_fit)
                        self.recent_xfitted.append(most_recent_xfitted)
                else:
                    self.recent_fit.append(self.current_fit)
                    self.recent_xfitted.append(self.current_xfitted)
            
            # calculates average fit and x values 
            self.recent_fit = self.recent_fit[-self.prev_frames:]            
            self.recent_xfitted = self.recent_xfitted[-self.prev_frames:]
            
            self.avg_fit = np.average(self.recent_fit, axis = 0)
            _, self.avg_xfitted = self.fit_quadratic(y_fitted=y_fitted,poly_fit=self.avg_fit)
            
            self.slope = 2*self.avg_fit[0]*y_fitted + self.avg_fit[1]
            
            return self.avg_xfitted, y_fitted
        
        else:
            self.detected = False
            self.false_detection_count += 1
            self.current_x = None  
            self.current_y = None
   
            return self.avg_xfitted, y_fitted

    def find_curvature(self,xcenters,ycenters,y_fitted):
    # finds curvature given (x,y) points and pixels to meters approximations    
        ym_per_pix = self.ym_per_pix 
        xm_per_pix = self.xm_per_pix
        curve_fit = np.polyfit(np.array(ycenters,np.float32)*ym_per_pix, np.array(xcenters,np.float32)*xm_per_pix,2) 
        self.curvature = ((1+(2*curve_fit[0]*y_fitted*ym_per_pix+curve_fit[1])**2)**1.5)/np.absolute(2*curve_fit[0])       
        return self.curvature

    
    
class Line_Set():
# defines a class to apply separation and slope filtering based on a pair of lane lines    
    def __init__(self):
        self.separation = None
        self.separation_limits_pct = [0.30,0.80]    # 0.20
        self.separation_diff_pct = 0.25
        self.slope_error_margin = 1
        self.slope_diff = None
    
    def sanity_check(self,left_lane,right_lane):
    # checks the difference in slope and the separation between the left and right lane candidates     
        if not(self.separation is None) and not(self.slope_diff is None):
            warp_size = left_lane.warp_size
            new_separation = np.array(right_lane.current_xfitted - left_lane.current_xfitted)
            separation_check1 = np.absolute(new_separation) >= self.separation_limits_pct[0]*warp_size[1]
            separation_check2 = np.absolute(new_separation) <= self.separation_limits_pct[1]*warp_size[1]
            separation_check = np.bitwise_and(separation_check1,separation_check2)
            new_slope_diff = np.array(right_lane.current_slope - left_lane.current_slope)
            slope_check = np.absolute(new_slope_diff) <= self.slope_error_margin                        
            if not(np.all(separation_check)) or not(np.all(slope_check)):
                # rejects one or both of lane candidates if separation or slope check fails
                if left_lane.detected == True and right_lane.detected == False:
                    left_lane = self.pop_line(left_lane)
                elif left_lane.detected == False and right_lane.detected == True:      
                    right_lane = self.pop_line(right_lane)
                elif left_lane.detected == True and right_lane.detected == True:      
                    right_lane = self.pop_line(right_lane)       
                    left_lane = self.pop_line(left_lane)                                                 
            elif np.all(new_separation > 0):
                self.separation = new_separation
                self.slope_diff = new_slope_diff                
        else:
            self.separation = np.array(right_lane.current_xfitted - left_lane.current_xfitted)
            self.slope_diff = np.array(right_lane.current_slope - left_lane.current_slope)        
        return left_lane, right_lane

    def pop_line(self,lane_class):
        # procedure to clean the lane class by dismissing the failed current lane candidate
        if lane_class.reset != True:
            lane_class.recent_fit.pop()
            lane_class.recent_xfitted.pop()
            lane_class.avg_fit = np.average(lane_class.recent_fit,axis=0)
            _, lane_class.avg_xfitted = lane_class.fit_quadratic(y_fitted=lane_class.yfitted,
                                                                 poly_fit=lane_class.avg_fit)
            lane_class.detected = False
            lane_class.false_detection_count += 2        
            print('popped!')
        
        return lane_class
          
```

```python
def abs_sobel_thresh(img, orient='x', sobel_kernel=3, thresh=(0, 255)):
    #finds thresholded binary sobel filtered image
    if len(img.shape) > 1:
        gray = np.mean(img, axis=2)
    else:
        gray = img
    if orient == 'x':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 1, 0))
    if orient == 'y':
        abs_sobel = np.absolute(cv2.Sobel(gray, cv2.CV_64F, 0, 1))
    scaled_sobel = np.uint8(255*abs_sobel/np.max(abs_sobel))
    binary_output = np.zeros_like(scaled_sobel)
    binary_output[(scaled_sobel >= thresh[0]) & (scaled_sobel <= thresh[1])] = 1
    return binary_output

def window_mask(width, height, img_ref, center,level):
    #draws windows below centroids
    output = np.zeros_like(img_ref)
    output[int(img_ref.shape[0]-level*height):int(img_ref.shape[0]-(level-1)*height),
           max(0,int(center-width/2)):min(int(center+width/2),img_ref.shape[1])] = 1
    return output

def region_of_interest(img, vertices):
    #masks images specified by convex region of interest
    mask = np.zeros_like(img)   
    if len(img.shape) > 2:
        channel_count = img.shape[2]  
        ignore_mask_color = (255,) * channel_count
    else:
        ignore_mask_color = 255  
    cv2.fillPoly(mask, vertices, ignore_mask_color)
    masked_image = cv2.bitwise_and(img, mask)
    return masked_image

def white_threshold(img,p=100,slack=30, hsv_range = 65,hls_range = 60):
    #detects white lanes based on adaptive thresholding and color masks (not used)
    masked_image = np.copy(img[:,:,2])
    masked_image[:int(0.50*img.shape[0]),:]=0
    percentile = int(np.percentile(masked_image,p)-slack)
    white_img1 = cv2.inRange(masked_image,percentile,255)
    mask1 = cv2.inRange(cv2.cvtColor(img, cv2.COLOR_RGB2HSV),np.array([0,0,255-hsv_range]), np.array([255,20,255]))
    mask2 = cv2.inRange(cv2.cvtColor(img, cv2.COLOR_RGB2HLS),np.array([0,255-hls_range,0]), np.array([255,255,hls_range]))
    mask3 = cv2.inRange(img, (200,200,200), (255,255,255))
    white_img2 = mask1 | mask2 | mask3
    binary_output = np.zeros_like(img[:,:,0])
    binary_output[(white_img1 == 255)|(white_img2==255)] = 1
    return binary_output

def yellow_threshold(img, p=100,slack=40):
    #detects yellow lanes based on adaptive thresholding and a color mask (not used)
    masked_image = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)[:,:,2]
    masked_image[:int(0.50*img.shape[0]),:]=0
    percentile = int(np.percentile(masked_image,p)-slack)
    yellow_img1 = cv2.inRange(masked_image,percentile,255)
    yellow_img2 = cv2.inRange(cv2.cvtColor(img, cv2.COLOR_RGB2HSV), np.array([20,50,50]),np.array([50,255,255]))
    binary_output = np.zeros_like(img[:,:,0])
    binary_output[(yellow_img1==255)|(yellow_img2==255)] = 1
    return binary_output


def process_images(folder,src,dst,vertices):
    #processes non-consecutive images in given folder
    images = glob.glob(folder+'test*.jpg')
    images = sorted(images, key=lambda string: int(string[-6:-4]))
    imgnum = [string[-6:-4] for string in images]
    global mtx, dist
    
    for idx, fname in zip(imgnum,images):
        original = mpimg.imread(fname)
        
        #apply distortion correction on original image
        img = cv2.undistort(original,mtx,dist,None,mtx)

        write_name = folder+'tracked'+idx+'.png'
        mpimg.imsave(write_name, img)
        
        #calculate sobel filtered binary composite images
        ycrcb = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
        yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
        hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
        hsv = cv2.cvtColor(img, cv2.COLOR_RGB2HSV)
        lab = cv2.cvtColor(img, cv2.COLOR_RGB2LAB)
        composite = np.stack((ycrcb[:, :, 0], ycrcb[:, :, 1], hls[:, :, 2]), axis=2)
        gradx = abs_sobel_thresh(composite, orient='x', thresh=(12,255))
        grady = abs_sobel_thresh(composite, orient='y', thresh=(25,255))

        #combined sobel x and y images
        preprocessImage = np.zeros_like(img[:,:,0])
        preprocessImage[((gradx==1)&(grady==1))] = 255

        binary_write_name = folder+'binary'+idx+'.png'
        mpimg.imsave(binary_write_name, preprocessImage, cmap = 'gray')
        
        #calculate and apply perspective transform
        M = cv2.getPerspectiveTransform(src, dst)
        Minv = cv2.getPerspectiveTransform(dst, src)
        warped = cv2.warpPerspective(preprocessImage, M, img_size, flags=cv2.INTER_LINEAR)
        warped_rgb = cv2.warpPerspective(img, M, img_size, flags=cv2.INTER_LINEAR)
        
        #plot transformation results
        if idx == imgnum[0]:
            gs = gridspec.GridSpec(1, 2)
            gs.update(wspace=0.2, hspace=0.2) 
            plt.figure(figsize=(16,9))
            
            srcx,srcy = zip(*src)
            dstx,dsty = zip(*dst)
            
            plt.subplot(1,3,1)
            plt.imshow(original.squeeze())
            plt.title('Orginal Image')
            
            plt.subplot(1,3,2)
            plt.imshow(img.squeeze())
            plt.title('Undistorted Image')
            plt.fill(srcx,srcy, edgecolor='r', fill=False)
            
            plt.subplot(1,3,3)
            plt.imshow(warped_rgb.squeeze())
            plt.title('Transformed Image')
            plt.fill(dstx,dsty, edgecolor='r', fill=False)

        #mask warped image through specified ROI points
        warped = region_of_interest(warped,vertices)
        
        warped_write_name = folder+'warped'+idx+'.png'
        mpimg.imsave(warped_write_name, warped, cmap = 'gray')
        
        #initialize lane class instances for every image 
        left_lane_class = Line(nature='Left')
        right_lane_class = Line(nature='Right')

        window_height = int(warped.shape[0]/left_lane_class.num_windows)
        kernel_width = left_lane_class.kernel_width

        #find lane centroids
        left_centroids = left_lane_class.find_window_centroids(warped)
        right_centroids = right_lane_class.find_window_centroids(warped)

        leftx,lefty = zip(*left_centroids) 
        rightx,righty = zip(*right_centroids)

        #define plot points
        yvals = range(int(warped.shape[0]*0), int(warped.shape[0]))
        
        l_points1 = np.zeros_like(warped)
        r_points1 = np.zeros_like(warped)
        l_points2 = np.zeros_like(warped)
        r_points2 = np.zeros_like(warped)
        
        window_height = int(warped.shape[0]/left_lane_class.num_windows)
        window_width = 2*left_lane_class.margin_pct*warped.shape[1]
       
        for level in range(1,left_lane_class.num_windows+1):
            l_mask1 = window_mask(left_lane_class.kernel_width, window_height, warped, leftx[level], level)
            r_mask1 = window_mask(right_lane_class.kernel_width, window_height, warped, rightx[level], level)
            l_points1[(l_points1 == 255) | (l_mask1 == 1)] = 255
            r_points1[(r_points1 == 255) | (r_mask1 == 1)] = 255
            l_mask2 = window_mask(window_width, window_height, warped, leftx[level-1], level)
            r_mask2 = window_mask(window_width, window_height, warped, rightx[level-1], level)
            l_points2[(l_points2 == 255) | (l_mask2 == 1)] = 255
            r_points2[(r_points2 == 255) | (r_mask2 == 1)] = 255
        
        #draw convolution windows
        template1 = np.array(r_points1+l_points1,np.uint8)
        template2 = np.array(r_points2+l_points2,np.uint8)
        zero_channel = np.zeros_like(template1)
        template = np.array(cv2.merge((zero_channel,template1,template2)),np.uint8)
        warpage = np.array(cv2.merge((warped,warped,warped)),np.uint8)
        windowed = cv2.addWeighted(warpage, 1, template, 0.5, 0.0)
        window_write_name = folder+'window'+idx+'.png'
        mpimg.imsave(window_write_name, windowed)
        
        #fit quadratic curves to centroids
        left_fit, left_fitx = left_lane_class.fit_quadratic(x_pixels=list(leftx),y_pixels=list(lefty),y_fitted=yvals)
        right_fit, right_fitx = right_lane_class.fit_quadratic(x_pixels=list(rightx),y_pixels=list(righty),y_fitted=yvals)   
        
        left_lane = np.array(list(zip(np.concatenate((left_fitx-kernel_width/2,left_fitx[::-1]+kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)
        right_lane = np.array(list(zip(np.concatenate((right_fitx-kernel_width/2,right_fitx[::-1]+kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)
        center_lane = np.array(list(zip(np.concatenate((left_fitx+kernel_width/2,right_fitx[::-1]-kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)
        
        #draw lane lines on warped image
        lane_image = np.zeros_like(img)
        cv2.fillPoly(lane_image,[left_lane],color=[255, 0 ,0])
        cv2.fillPoly(lane_image,[right_lane],color=[0,0,255])
        lanes_binary = cv2.addWeighted(warpage, 1, lane_image, 0.75, 0.0)
        lanes_write_name = folder+'lanes'+idx+'.png'
        mpimg.imsave(lanes_write_name, lanes_binary)
        
        road = np.zeros_like(img)
        road_bkg = np.zeros_like(img)
        cv2.fillPoly(road,[left_lane],color=[255, 0 ,0])
        cv2.fillPoly(road,[right_lane],color=[0,0,255])
        cv2.fillPoly(road,[center_lane],color=[30,144,255])
        cv2.fillPoly(road_bkg,[left_lane],color=[255,255,255])
        cv2.fillPoly(road_bkg,[right_lane],color=[255,255,255])
        
        # take inverse transform of warped image
        road_warped = cv2.warpPerspective(road,Minv,img_size,flags=cv2.INTER_LINEAR)
        road_warped_bkg = cv2.warpPerspective(road_bkg,Minv,img_size,flags=cv2.INTER_LINEAR)
        
        #draw lane lines on undistorted image
        base = cv2.addWeighted(img, 1.0, road_warped_bkg, -1.0, 0.0)
        result = cv2.addWeighted(base, 1.0, road_warped, 1.0, 0.0)

        ym_per_pix = left_lane_class.ym_per_pix
        xm_per_pix = left_lane_class.xm_per_pix
        
        #calculate curvature and distance from center
        curverad_left = left_lane_class.find_curvature(leftx,lefty,yvals)[-1]
        curverad_right = right_lane_class.find_curvature(rightx,righty,yvals)[-1]
        camera_center = (left_fitx[-1] + right_fitx[-1])/2
        center_diff = (camera_center-warped.shape[1]/2)*xm_per_pix
        side_pos = 'left'
        if center_diff <=0:
            side_pos = 'right'
        
        #display curvature and lateral position
        cv2.putText(result,'Left Lane Radius of Curvature = '+str(round(curverad_left,3))+'(m)',(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
        cv2.putText(result,'Right Lane Radius of Curvature = '+str(round(curverad_right,3))+'(m)',(50,100),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
        cv2.putText(result,'Vehicle is '+str(abs(round(center_diff,3)))+'m '+side_pos+' of center',(50,150), cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)


        write_name = folder+'tracked'+idx+'.png'
        mpimg.imsave(write_name, result)

img = np.zeros((720,1280,3))        

#specify ROI vertices for 'project' video
vertices1 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
                    (0.90*img.shape[1],0),(0.90*img.shape[1],img.shape[0])]], dtype=np.int32) 
    
#calculate source and destination points for perspective transform    
bot_width = 0.60
mid_width = 0.08
height_pct = 0.62
bottom_trim = 0.95
offset = img.shape[1]*0.25
    
src1 = np.float32([[img.shape[1]*(0.5-mid_width/2), img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+mid_width/2), img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+bot_width/2), img.shape[0]*bottom_trim],
                  [img.shape[1]*(0.5-bot_width/2), img.shape[0]*bottom_trim]])
 
dst1 = np.float32([[offset,0],
                  [img.shape[1]-offset, 0],
                  [img.shape[1]-offset, img.shape[0]],
                  [offset, img.shape[0]]])

process_images('./test_images/test_images1/',src1,dst1,vertices1)

#specify ROI vertices for 'challenge' video
vertices2 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
#                       (0.45*img.shape[1],0),(0.425*img.shape[1],img.shape[0]),
#                       (0.575*img.shape[1],img.shape[0]),(0.55*img.shape[1],0),
                      (0.85*img.shape[1],0),(0.825*img.shape[1],img.shape[0])]], dtype=np.int32) 

#calculate source and destination points for perspective transform
bot_width = 0.50
mid_width = 0.08
height_pct = 0.65
bottom_trim = 0.95
offset = img.shape[1]*0.25
shift = img.shape[1]/32
    
src2 = np.float32([[img.shape[1]*(0.5-mid_width/2)+shift, img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+mid_width/2)+shift, img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+bot_width/2)+shift, img.shape[0]*bottom_trim],
                  [img.shape[1]*(0.5-bot_width/2)+shift, img.shape[0]*bottom_trim]])
 
dst2 = np.float32([[offset,0],
                  [img.shape[1]-offset, 0],
                  [img.shape[1]-offset, img.shape[0]],
                  [offset, img.shape[0]]])

process_images('./test_images/test_images2/',src2,dst2,vertices2)


#specify ROI vertices for 'harder challenge' video
vertices3 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
                      (0.45*img.shape[1],0),(0.425*img.shape[1],img.shape[0]),
                      (0.575*img.shape[1],img.shape[0]),(0.55*img.shape[1],0),
                      (0.85*img.shape[1],0),(0.825*img.shape[1],img.shape[0])]], dtype=np.int32)  

#calculate source and destination points for perspective transform
bot_width = 0.50
mid_width = 0.18   
height_pct = 0.68  
bottom_trim = 0.95
offset = img.shape[1]*0.25
    
src3 = np.float32([[img.shape[1]*(0.5-mid_width/2), img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+mid_width/2), img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+bot_width/2), img.shape[0]*bottom_trim],
                  [img.shape[1]*(0.5-bot_width/2), img.shape[0]*bottom_trim]])
 
dst3 = np.float32([[offset,0],
                  [img.shape[1]-offset, 0],
                  [img.shape[1]-offset, img.shape[0]],
                  [offset, img.shape[0]]])

process_images('./test_images/test_images3/',src3,dst3,vertices3)
```

**2. Creation of binary thresholded image.**


I used a composite image whose three components were taken respectively from the 'Y' Luma component of the YCrCb colorspace, the 'U' chrominance component of the YUV colorspace and the the 'S' saturation component of the HLS colorspace. The Luma component gives good detection of white lanes, the chrominance commponents is suitable for yellow lane dection while the saturation components leads to robustness in shadows and high exposure parts of the image. Using Sobel filters in the x and y directions on the composite grayscale image and then thresholding the result was key for robust detection even in the very noisy 'harder challenge' video. **3. Perspective Transform**


A perspective transform was obtained from source and destination points in order to make the lanes roughly parallel. The perspective transforms although hardcoded, were slightly different for each video representing the change in the road area in each video. Perspective transforms have been handled in the code cell above. Source and destination points are drawn in the second and third column of the output of the code cell. A bird's eye view is captured successfully most of the times and the lane detection algorithm yields better results on the warped  binary image i.e. the image obtained after applying the perspective transform on the binary image obtained in the previous step. 


**4. Polynomial fitting**


After obtaining a warped binary image, a convolution or sliding window approach was used to identify the lane centroids for both left and right lanes. This approach entailed convolving a Gaussian kernel over the vertical sum of pixels values lying along a horizontal slice of the image and determining where the maximum value occured for that slice. The slices were trimmed horizontally to a margin around below/previous centroids for each lane. After lane centroids are found, a quadratic curve is fitted to those points using the least squared error method. A sample of the results corresponding to the three test images considered previously are shown below. The blue regions represent the convolution windows while the green bars represents the the pixels contributing to the maximum convolved sum in the blue window (the lane centroids are located at the top center of each green bar). Least square fit left and right curves are also drawn in corresponding images. (Note that ROI masks have been used on the binary warped image to reduce the effect of left/right border areas)

**5. Radius of curvature and center of lane**

The radius of curvature is found out by first fitting a quadratic curve to (x,y) points which have been converted to meters using approximate relations between lengths in pixels and meters. Then the radius of curvature formula has been utilized to calculate respective curvatures for left and right lanes. The position of the vehicle with respect to the center is calculated by calculating the midpoint between detected left and right lanes and the offeset of the center of the image from that point. Distances are reported in meters as before and the camera is assumed to be placed at the center of the car. Parts of code where these calculations have been done are noted in the code.

```python
from moviepy.editor import VideoFileClip
from IPython.display import HTML
import random

def process_video(img,src,dst,vertices):
    #processes consecutive frames one frame at a time
    global mtx, dist
    
    img = cv2.undistort(img,mtx,dist,None,mtx)
    img_size = (img.shape[1], img.shape[0])
    
    ycrcb = cv2.cvtColor(img, cv2.COLOR_RGB2YCrCb)
    yuv = cv2.cvtColor(img, cv2.COLOR_RGB2YUV)
    hls = cv2.cvtColor(img, cv2.COLOR_RGB2HLS)
    composite = np.stack((ycrcb[:, :, 0], yuv[:, :, 1], hls[:, :, 2]), axis=2)
    gradx = abs_sobel_thresh(composite, orient='x', thresh=(12,255))
    grady = abs_sobel_thresh(composite, orient='y', thresh=(25,255))
    preprocessImage = np.zeros_like(img[:,:,0])
    preprocessImage[((gradx==1)&(grady==1))] = 255

    img_size = (img.shape[1], img.shape[0])
    
    M = cv2.getPerspectiveTransform(src, dst)
    Minv = cv2.getPerspectiveTransform(dst, src)
    warped = cv2.warpPerspective(preprocessImage, M, img_size, flags=cv2.INTER_LINEAR)
    warped = region_of_interest(warped,vertices)

    #invoke global lane class objects to detect, store and track lanes across frames 
    global left_lane_class, right_lane_class, lane_pair_class
    
    window_height = int(warped.shape[0]/left_lane_class.num_windows)
    kernel_width = left_lane_class.kernel_width
    
    left_centroids = left_lane_class.find_window_centroids(warped)
    right_centroids = right_lane_class.find_window_centroids(warped)
    
    yvals = range(int(warped.shape[0]*0), int(warped.shape[0]))
            
    leftx = [x for x,y in left_centroids]
    lefty = [y for x,y in left_centroids]
    
    rightx = [x for x,y in right_centroids]
    righty = [y for x,y in right_centroids]

    left_lane_class.tracking_operations(leftx,lefty,yvals)
    right_lane_class.tracking_operations(rightx,righty,yvals)
    
    #perform sanity check - separation and slope check on the pair of lines detected
    left_lane_class,right_lane_class = lane_pair_class.sanity_check(left_lane_class,right_lane_class)

    _, left_fitx = left_lane_class.fit_quadratic(y_fitted=yvals,poly_fit=left_lane_class.avg_fit)
    _, right_fitx = right_lane_class.fit_quadratic(y_fitted=yvals,poly_fit=right_lane_class.avg_fit)
    
    left_lane = np.array(list(zip(np.concatenate((left_fitx-kernel_width/2,left_fitx[::-1]+kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)
    right_lane = np.array(list(zip(np.concatenate((right_fitx-kernel_width/2,right_fitx[::-1]+kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)
    center_lane = np.array(list(zip(np.concatenate((left_fitx+kernel_width/2,right_fitx[::-1]-kernel_width/2), axis=0),np.concatenate((yvals,yvals[::-1]),axis=0))),np.int32)

    road = np.zeros_like(img)
    road_bkg = np.zeros_like(img)
    cv2.fillPoly(road,[left_lane],color=[255, 0 ,0])
    cv2.fillPoly(road,[right_lane],color=[0,0,255])
    cv2.fillPoly(road,[center_lane],color=[30,144,255])
    cv2.fillPoly(road_bkg,[left_lane],color=[255,255,255])
    cv2.fillPoly(road_bkg,[right_lane],color=[255,255,255])

    road_warped = cv2.warpPerspective(road,Minv,img_size,flags=cv2.INTER_LINEAR)
    road_warped_bkg = cv2.warpPerspective(road_bkg,Minv,img_size,flags=cv2.INTER_LINEAR)

    base = cv2.addWeighted(img, 1.0, road_warped_bkg, -1.0, 0.0)
    result = cv2.addWeighted(base, 1.0, road_warped, 1.0, 0.0)

    ym_per_pix = left_lane_class.ym_per_pix
    xm_per_pix = left_lane_class.xm_per_pix
    
    _, leftx_avg = left_lane_class.fit_quadratic(y_fitted=np.array(lefty,np.int32),poly_fit=left_lane_class.avg_fit)
    _, rightx_avg = right_lane_class.fit_quadratic(y_fitted=np.array(lefty,np.int32),poly_fit=right_lane_class.avg_fit)

    curverad_left = left_lane_class.find_curvature(leftx_avg,lefty,yvals)[-1]
    curverad_right = right_lane_class.find_curvature(rightx_avg,righty,yvals)[-1]
    
    camera_center = (left_fitx[-1] + right_fitx[-1])/2
    center_diff = (camera_center-warped.shape[1]/2)*xm_per_pix
    side_pos = 'left'
    if center_diff <=0:
        side_pos = 'right'

    cv2.putText(result,'Left Lane Radius of Curvature = '+str(round(curverad_left,3))+'(m)',(50,50),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
    cv2.putText(result,'Right Lane Radius of Curvature = '+str(round(curverad_right,3))+'(m)',(50,100),cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
    cv2.putText(result,'Vehicle is '+str(abs(round(center_diff,3)))+'m '+side_pos+' of center',(50,150), cv2.FONT_HERSHEY_SIMPLEX,1,(255,255,255),2)
    
    return result
 
def wrapper_project(img):
    #wrapper function for project video
    
    global left_lane_class, right_lane_class, lane_pair_class
    
    vertices1 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
                    (0.90*img.shape[1],0),(0.90*img.shape[1],img.shape[0])]], dtype=np.int32) 
    
    bot_width = 0.60
    mid_width = 0.08
    height_pct = 0.62
    bottom_trim = 0.95
    offset = img.shape[1]*0.25
    shift = 0
    
    src1 = np.float32([[img.shape[1]*(0.5-mid_width/2)+shift, img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+mid_width/2)+shift, img.shape[0]*height_pct],
                  [img.shape[1]*(0.5+bot_width/2)+shift, img.shape[0]*bottom_trim],
                  [img.shape[1]*(0.5-bot_width/2)+shift, img.shape[0]*bottom_trim]])
 
    dst1 = np.float32([[offset,0],
                  [img.shape[1]-offset, 0],
                  [img.shape[1]-offset, img.shape[0]],
                  [offset, img.shape[0]]])
    
    
    result = process_video(img, src1, dst1, vertices1)
    
    return result   
    
def wrapper_challenge(img):
    #wrapper function for challenge video
    
    global left_lane_class, right_lane_class, lane_pair_class
    
    
    vertices2 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
                          (0.85*img.shape[1],0),(0.825*img.shape[1],img.shape[0])]], dtype=np.int32) 
    
    bot_width = 0.50
    mid_width = 0.08
    height_pct = 0.65
    bottom_trim = 0.95
    offset = img.shape[1]*0.25
    shift = img.shape[1]/32

    src2 = np.float32([[img.shape[1]*(0.5-mid_width/2)+shift, img.shape[0]*height_pct],
                      [img.shape[1]*(0.5+mid_width/2)+shift, img.shape[0]*height_pct],
                      [img.shape[1]*(0.5+bot_width/2)+shift, img.shape[0]*bottom_trim],
                      [img.shape[1]*(0.5-bot_width/2)+shift, img.shape[0]*bottom_trim]])

    dst2 = np.float32([[offset,0],
                      [img.shape[1]-offset, 0],
                      [img.shape[1]-offset, img.shape[0]],
                      [offset, img.shape[0]]])
    
    
    result = process_video(img, src2, dst2, vertices2)
    
    return result 

def wrapper_harder_challenge(img):
    ##wrapper function for harder challenge video
    
    global left_lane_class, right_lane_class, lane_pair_class
    
    lane_pair_class = Line_Set() #comment this line for alternative video
    
    #masking for harder challenge video  
    vertices3 = np.array([[(0.10*img.shape[1],img.shape[0]),(0.10*img.shape[1],0),
                      (0.45*img.shape[1],0),(0.425*img.shape[1],img.shape[0]),
                      (0.575*img.shape[1],img.shape[0]),(0.55*img.shape[1],0),
                      (0.85*img.shape[1],0),(0.825*img.shape[1],img.shape[0])]], dtype=np.int32) 

    #alternative masking for harder challenge video
#     vertices3 = np.array([[(0.10*img.shape[1],img.shape[0]),(0*img.shape[1],0),
#                           (0.45*img.shape[1],0),(0.425*img.shape[1],img.shape[0]),
#                           (0.575*img.shape[1],img.shape[0]),(0.55*img.shape[1],0),
#                           (1*img.shape[1],0),(0.90*img.shape[1],img.shape[0])]], dtype=np.int32)
    
    #implements a dynamic horizontal shift in the perspective transform to map more lane area at turns
    if not(left_lane_class.avg_fit is None) and not(right_lane_class.avg_fit is None):       
        left_shift = 0.10*img.shape[1]-left_lane_class.avg_xfitted[0]
        right_shift = right_lane_class.avg_xfitted[0] - 0.90*img.shape[1]
        shift = right_shift*(np.sign(right_shift)+1)*0.5 - left_shift*(np.sign(left_shift)+1)*0.5 
    else:
        shift = 0
    
    bot_width = 0.50
    mid_width = 0.18   
    height_pct = 0.68  
    bottom_trim = 0.95
    offset = img.shape[1]*0.25
    
    src3 = np.float32([[max(min(img.shape[1]*(0.5-mid_width/2)+shift,img.shape[1]),0), img.shape[0]*height_pct],
                     [max(min(img.shape[1]*(0.5+mid_width/2)+shift,img.shape[1]),0), img.shape[0]*height_pct],
                     [max(min(img.shape[1]*(0.5+bot_width/2)+shift,img.shape[1]),0), img.shape[0]*bottom_trim],
                     [max(min(img.shape[1]*(0.5-bot_width/2)+shift,img.shape[1]),0), img.shape[0]*bottom_trim]])
 
    dst3 = np.float32([[offset,0],
                     [img.shape[1]-offset, 0],
                     [img.shape[1]-offset, img.shape[0]],
                     [offset, img.shape[0]]])
    
    
    result = process_video(img, src3, dst3, vertices3)
    
    return result                          


    
```

## Discussion

This was a difficult project - running the pipeline on the relatively easier 'project' was relatively straightforward as discussed before but the the two challenge videos especially the harder challenge video required rethinking and tweaking the solution. The convolution approach as previously outlined was used to find out lane centroids. Filtering of lanes using variations in slope, separation, polynomial coefficients and x-displacements was done by creating appropriate lane class objects to reject outliers. Lane centroids were searched in regions around previous high confidence detections. In first few attempts color transforms were directly used by thresholding and worked well for first two videos except the challenge video where it lead to an extremely noisy image due to the proximity of the offroad vegetation and other structures. Thereafter, only edges were detected using Sobel filters in x and y directions, that too on a composite image composed of channels taken from the YCrCb, YUV and HSV colorspaces. This proved to be a critical choice for the last video as it provided some resilience in the presence of shadows. Another problem encountered in the challenge video is that the lanes constantly fall outside the domain of the perspective transform leading to no detection of lanes. This was later tackled using a dynamic horizontal shift of perspective based on the distance of lanes detected in the previous frames from the top corners of the warped binary image. (Checkout the alternative harder challenge video to see its full effect). Lastly, for the harder challenge the tracking needed to be reset after a certain number of false detections due to highly dynamic lanes. This was implemented by maintaining a false detection count. Despite all these measures, the binary images were still noisy on the edges due to presence of walls, bridges, car etc. So a Region of interest mask was applied in all videos. In the end, a decent result was obtained on all three videos by stretching the capabilities of the convolution approach.

For further work, I might use CNNs to generate binary images so that the network learns which edges are important and which aren't. The detection and tracking framework should still be used with CNNs as then parameters like distance from the lane center and lane curvature can be easily calculated. An end-to-end CNN architecture which colors the lane area should be avoided as in my opinion it would be difficult to track lanes across frames and reject outliers.
